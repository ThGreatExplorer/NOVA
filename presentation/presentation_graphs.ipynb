```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# --- Simulation Parameters ---
dt = 0.01
T = 10
n_steps = int(T / dt)
t = np.linspace(0, T, n_steps)

# Black hole observation parameters
mean_level = 0.5            # the average flux/value
noise_std = 0.2             # how noisy the data is
drift_strength = 0.02        # optional slow drift (set to 0 for none)

# Optional slow drift (e.g., gradual accretion changes)
drift = drift_strength * np.sin(0.1 * t)  # very slow variation

# Generate normally distributed observations around the mean + drift
observations = mean_level + drift + np.random.normal(0, noise_std, n_steps)

# --- Visualization ---
fig, ax = plt.subplots()
ax.tick_params(axis='both', which='both', bottom=False, top=False,
               left=False, right=False, labelbottom=False, labelleft=False)

ax.plot(t, observations, alpha=0.8)

# plt.xlabel("Time (s)")
# plt.ylabel("Observation Value")
plt.title("Expected Black Hole Observation")
plt.show()
```

```python
# --- Simulation Parameters ---
dt = 0.01
T = 10
n_steps = int(T / dt)
t = np.linspace(0, T, n_steps)

# Pulsar-like sinusoid parameters
frequency = 1.0       # Hz (controls the pulsation)
amplitude = 1.0       # main signal amplitude
decay_rate = 0.0005   # exponential decay factor (optional, set to 0 to disable)

# Noise parameters
noise_strength = 0.3  # standard deviation of the Gaussian noise

# pure sinusoid
signal = amplitude * np.sin(2 * np.pi * frequency * t)

# optional exponential decay (pulsars often weaken)
signal *= np.exp(-decay_rate * t)

# add Gaussian noise
noise = np.random.normal(0, noise_strength, size=n_steps)
noisy_signal = signal + noise

# --- Visualization ---
fig, ax = plt.subplots()
ax.plot(t, noisy_signal, alpha=0.7)
ax.tick_params(axis='both', which='both', bottom=False, top=False,
               left=False, right=False, labelbottom=False, labelleft=False)

# plt.xlabel("Time (s)")
# plt.ylabel("Value")
plt.title("Expected Pulsar Observation")
plt.show()
```
